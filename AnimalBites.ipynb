# Feature Engineering

```python
AnimalBites['Male'] = (AnimalBites['GenderIDDesc'] == 'MALE').astype(int)
AnimalBites['Female'] = (AnimalBites['GenderIDDesc'] == 'FEMALE').astype(int)

print(AnimalBites.head())
```

```python
is_datetime = isinstance(AnimalBites['bite_date'], pd.DatetimeIndex)
print(is_datetime)
```

```python
AnimalBites['bite_date'] = pd.to_datetime(AnimalBites['bite_date'], errors='coerce')

invalid_dates = AnimalBites[AnimalBites['bite_date'].isna()]['bite_date']
if not invalid_dates.empty:
    print(f"Invalid datetime values found:\n{invalid_dates}")
    
AnimalBites['Date'] = AnimalBites['bite_date'].dt.date
AnimalBites['Time'] = AnimalBites['bite_date'].dt.time

print(AnimalBites.head()) 
```

```python
# Filter out rows with NaT in 'bite_date' column
AnimalBites = AnimalBites[AnimalBites['bite_date'].notna()]

AnimalBites['bite_date'] = pd.to_datetime(AnimalBites['bite_date'])
AnimalBites['Date'] = AnimalBites['bite_date'].dt.date
AnimalBites['Time'] = AnimalBites['bite_date'].dt.time

print(AnimalBites.head())
```

```python
AnimalBites['bite_date'] = pd.to_datetime(AnimalBites['bite_date'], errors='coerce')

AnimalBites['Year'] = AnimalBites['bite_date'].dt.year
AnimalBites['Month'] = AnimalBites['bite_date'].dt.month
AnimalBites['Day'] = AnimalBites['bite_date'].dt.day

print(AnimalBites.head())
```

# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

years = sorted(AnimalBites['Year'].unique())
years = [year for year in years if 1950 <= year <= 2022] 

yearly_counts = AnimalBites['Year'].value_counts().sort_index()

plt.figure(figsize=(14, 8)) 

bar_width = 0.8
index = yearly_counts.index  
positions = range(len(index))  

plt.bar(positions, yearly_counts.values, width=bar_width, color='skyblue', align='center')

plt.title('Number of Victims per Year (1950 - 2022)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Victims', fontsize=14)
plt.xticks(positions, index, rotation=45, fontsize=10) 
plt.yticks(fontsize=10)  

for i, v in enumerate(yearly_counts.values):
    plt.text(i, v + 50, str(v), ha='center', va='bottom', fontsize=8)

plt.grid(False)

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=AnimalBites, x='GenderIDDesc', palette='Set2')
plt.title('Distribution of Victims by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Victims')
plt.show()
```

```python
filtered_data = AnimalBites[(AnimalBites['Year'] >= 1950) & (AnimalBites['Year'] <= 2022)]

yearly_vaccinated_counts = filtered_data[filtered_data['vaccination_yrs'] > 0]['Year'].value_counts().sort_index()

plt.figure(figsize=(14, 8)) 

bar_width = 0.8

positions = range(len(yearly_vaccinated_counts.index))

plt.bar(positions, yearly_vaccinated_counts.values, width=bar_width, color='orange')

plt.title('Number of Vaccinated Victims per Year (1950 - 2022)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(positions, yearly_vaccinated_counts.index, rotation=45, fontsize=10) 
plt.yticks(fontsize=10) 

plt.grid(False) 

plt.show()
```

```python
filtered_data = AnimalBites[['vaccination_yrs', 'victim_zip']].dropna()

filtered_data['victim_zip'] = pd.to_numeric(filtered_data['victim_zip'], errors='coerce').astype('Int64')

filtered_data = filtered_data.dropna()

plt.figure(figsize=(10, 8))

sns.scatterplot(x='vaccination_yrs', y='victim_zip', data=filtered_data)

plt.title('Relationship between Vaccination Years and Victim ZIP Code')
plt.xlabel('Vaccination Years')
plt.ylabel('Victim ZIP Code')

plt.show()
```
